```python
import pandas as pd
import requests

url = "https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)"

# Send a request with a browser-like header
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0 Safari/537.36"}
response = requests.get(url, headers=headers)
response.raise_for_status()

# Parse all tables
tables = pd.read_html(response.text)
pop = tables[0]
pop_na = pop[pop['UN statistical subregion[1]'] == 'Northern America']
```

```python
pop_na
```

```python
# merge the happiness scores and population data together
happy_scores.merge(pop, how='inner', left_on = 'country_name', right_on='Country or territory')
```

```python
# outer join
happy_scores.merge(pop, how='outer', left_on = 'country_name', right_on='Country or territory')
```

```python
# left join
happy_scores.merge(pop, how='left', left_on = 'country_name', right_on='Country or territory')
```
